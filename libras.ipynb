# ***Logistic Regression***

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
scaler = StandardScaler()
X=libras_filtered.drop('Unnamed: 90',axis=1)
y=libras_filtered['Unnamed: 90']

libras_scaled = scaler.fit_transform(libras_filtered)

columns_to_leave_unchanged = ['Unnamed: 90']

columns_to_scale = [col for col in libras_filtered.columns if col not in columns_to_leave_unchanged]

preprocessor = ColumnTransformer(
    transformers=[("scale", StandardScaler(),X.columns)],
    remainder="passthrough",
)

X_scaled_ = preprocessor.fit_transform(X)

columns_after_scaling = columns_to_scale 

X_scaled = pd.DataFrame(X_scaled_, columns=columns_to_scale)

X_train, X_test, y_train, y_test = train_test_split( X_scaled, y, test_size=0.3, random_state=101)
```

```python
# Create a logistic regression model
model = LogisticRegression(max_iter=1000)

# Train the model on the training set
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Print the results
print(f'Accuracy: {accuracy:.2f}')
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Confusion Matrix:\n{conf_matrix}')
```

```python
classification_rep = classification_report(y_test, y_pred)
print(f'Classification Report:\n{classification_rep}')
```

```python
cm_test = confusion_matrix(y_test, y_pred)
ax = plt.subplot()
sns.heatmap(cm_test, annot=True, fmt="d", cmap="Blues")
ax.set_xlabel("Predicted labels")
ax.set_ylabel("True labels")
ax.set_title("Confusion Matrix")
plt.show()
```

# ***KNN ANALYSIS OF DATASET***

```python
pred_classifier=[]
pred_accuracy=[]
error_rate=[]

X_train, X_test, y_train, y_test = train_test_split( X_scaled.values, y.values, test_size=0.3, random_state=101)

for i in range(2,15):
    classifier = KNeighborsClassifier(n_neighbors = i) 
    classifier.fit(X_train, y_train) 
    
    prediction_i=classifier.predict(X_test)
    error_rate.append(np.mean(prediction_i != y_test))

    cm=confusion_matrix(y_test,prediction_i)
    print(f'Confusion Matrix for k={i} : ')
    
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cbar=False)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title(f'Confusion Matrix')
    plt.show()

    pred_accuracy.append(accuracy_score(y_test,prediction_i))
```

```python
for i in range(1,13):
  print("The accuracy for KNN model for k = {}".format(i+1))
  print(pred_accuracy[i-1])
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,15),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
